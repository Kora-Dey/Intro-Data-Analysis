```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
from pandas import set_option
from pandas import read_csv
import numpy as np

colnames = ['Outout','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']

filename = 'wine.csv'
data = read_csv(filename,header=None, names=colnames )

data.head()
```

```python
data['Proline_logtransform'] = np.log10(data['Proline'])
```

```python
data
```

```python
feature_name ='Ash'
max_value = data[feature_name].max()
min_value = data[feature_name].min()
data[feature_name] = (data[feature_name] - min_value) / (max_value - min_value)
```
