```python
#Import Python Libraries

import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# load boston data set. If the dataset is not in the current directory, you will need to provide
# a full path to filename variable. Check the dataframe after loading to determine if there is any extra columns,
# if there is eliminate the extra column. The sample code for loading dataset from current directory is given.
# Boston dataset column tags
names = ['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','black','lstat','medv']

filename = 'boston.csv'
data = pd.read_csv(filename)[names]
```

```python
# perform data scaling by normalizing only the X (we don't normally perform transformation on the Y/output)

############
# to do
############

X_norm = X.copy()
cols = X_norm.columns.to_list()

for column in cols:
    X_norm[column] = (X_norm[column] - X_norm[column].min()) / (X_norm[column].max() - X_norm[column].min())    

    
```

```python
# calculate the descriptive statistics after normalization

############
# to do
############

X_norm.describe()
```

```python
# plot histogram of X and compare with histograms before normalization.
# Does normalization improves the data for predictive modeling?

############
# to do
############

X_norm.hist(figsize=(15, 20))


```

```python
# perform data scaling by standardizing only the X (we don't normally perform transformation on the Y/output)

############
# to do
############

X_standard = X.copy()
cols = X_standard.columns.to_list()

for column in cols:
    X_standard[column] = (X_standard[column] - X_standard[column].mean()) / X_standard[column].std()

```

```python
# calculate the descriptive statistics after standardization

############
# to do
############

X_standard.describe()
```

```python
# plot histogram of X and compare with histograms before standardizing.
# Does standardizing improves the data for predictive modeling?

############
# to do
############

X_standard.hist(figsize=(15, 20))
```

```python
# plot histogram of X and compare with histograms before standardizing.
# Does standardizing improves the data for predictive modeling?

## Ploting all 3 histograms side by side - raw value, normalized value and standardized value.
##

column = 'tax'

plt.rcParams["figure.figsize"] = [12.50, 3.50]
plt.rcParams["figure.autolayout"] = True

fig, axes = plt.subplots(1, 3)

X.hist(column, bins=10, ax=axes[0])
X_norm.hist(column, bins=10, ax=axes[1])
X_standard.hist(column, bins=10, ax=axes[2])

plt.show()

## It looks the distribution or shape does not change due to  normalization or standardization. It changes the x-axis ranges.
# This is true for all columns
# Having standardized data helps converging quickly. 
# If features are on  almost the same scale i.e., equally important and  regression would be easier to process and converge quickly
```
